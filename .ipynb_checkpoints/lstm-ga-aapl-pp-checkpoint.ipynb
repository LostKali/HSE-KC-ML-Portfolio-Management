```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input

# Импортируем библиотеки для генетического алгоритма
from deap import base, creator, tools, algorithms

# Загрузка данных
data = pd.read_csv('sp500_historical_data.csv')

# Фильтруем данные для одной акции
ticker = 'AAPL'  # Замените на нужный тикер
df = data[data['Ticker'] == ticker]

# Сортируем данные по дате
df = df.sort_values('Date')

# Преобразуем столбец 'Date' в формат datetime и устанавливаем его в качестве индекса
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Выбираем столбец 'Close' для прогнозирования
prices = df[['Close']]

# Нормализация данных
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(prices)

# Разделение данных на обучающую и тестовую выборки
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[0:train_size]
test_data = scaled_data[train_size:]

# Создание обучающих и тестовых наборов данных
def create_dataset(dataset, time_step):
    X, Y = [], []
    for i in range(len(dataset) - time_step -1):
        X.append(dataset[i:(i + time_step), 0])
        Y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(Y)

# Функция для создания и оценки модели
def evaluate_model(individual):
    time_step = int(individual[0])
    neurons = int(individual[1])
    batch_size = int(individual[2])
    epochs = int(individual[3])

    # Проверяем, что batch_size не больше размера данных
    if batch_size > len(train_data):
        batch_size = len(train_data)

    X_train, y_train = create_dataset(train_data, time_step)
    X_test, y_test = create_dataset(test_data, time_step)

    # Проверяем, что размер данных позволяет сформировать наборы
    if len(X_train) == 0 or len(X_test) == 0:
        return (float('inf'),)  # Возвращаем бесконечную ошибку, чтобы отбросить эту комбинацию гиперпараметров

    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    model = Sequential()
    model.add(Input(shape=(time_step, 1)))
    model.add(LSTM(units=neurons, return_sequences=True))
    model.add(LSTM(units=neurons))
    model.add(Dense(1))

    model.compile(loss='mean_squared_error', optimizer='adam')

    # Обучаем модель
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

    # Прогнозируем на тестовых данных
    test_predict = model.predict(X_test)
    test_predict = scaler.inverse_transform(test_predict)
    y_test_actual = scaler.inverse_transform([y_test])

    # Рассчитываем RMSE
    rmse = math.sqrt(mean_squared_error(y_test_actual[0], test_predict[:,0]))

    # Возвращаем метрику (чем меньше, тем лучше)
    return (rmse,)

# Настройка генетического алгоритма
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))  # Минимизируем RMSE
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()

# Определяем диапазоны гиперпараметров
toolbox.register("time_step", np.random.randint, 10, 100)      # Размер окна (time_step)
toolbox.register("neurons", np.random.randint, 10, 100)        # Количество нейронов
toolbox.register("batch_size", np.random.randint, 16, 128)     # Размер пакета
toolbox.register("epochs", np.random.randint, 5, 20)           # Количество эпох

toolbox.register("individual", tools.initCycle, creator.Individual,
                 (toolbox.time_step, toolbox.neurons, toolbox.batch_size, toolbox.epochs), n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", evaluate_model)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutUniformInt, low=[10, 10, 16, 5], up=[100, 100, 128, 20], indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

# Запуск генетического алгоритма
def run_ga():
    population = toolbox.population(n=10)
    ngen = 5  # Количество поколений

    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("min", np.min)
    stats.register("avg", np.mean)

    hof = tools.HallOfFame(1)  # Сохраняем лучший результат

    algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=ngen,
                        stats=stats, halloffame=hof, verbose=True)

    return hof[0]

best_individual = run_ga()
print(f"Лучшие гиперпараметры: time_step={int(best_individual[0])}, neurons={int(best_individual[1])}, batch_size={int(best_individual[2])}, epochs={int(best_individual[3])}")

# Обучаем модель с лучшими гиперпараметрами
time_step = int(best_individual[0])
neurons = int(best_individual[1])
batch_size = int(best_individual[2])
epochs = int(best_individual[3])

X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

model = Sequential()
model.add(Input(shape=(time_step, 1)))
model.add(LSTM(units=neurons, return_sequences=True))
model.add(LSTM(units=neurons))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1)

# Прогнозирование и оценка модели
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Обратное масштабирование предсказаний
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train_actual = scaler.inverse_transform([y_train])
y_test_actual = scaler.inverse_transform([y_test])

# Расчет RMSE
train_rmse = math.sqrt(mean_squared_error(y_train_actual[0], train_predict[:, 0]))
test_rmse = math.sqrt(mean_squared_error(y_test_actual[0], test_predict[:, 0]))

print(f'RMSE на обучающей выборке: {train_rmse}')
print(f'RMSE на тестовой выборке: {test_rmse}')

# Построение графика предсказанных и реальных цен
plt.figure(figsize=(12, 6))
plt.plot(df.index[time_step:len(train_predict)+time_step], y_train_actual[0], label='Реальная цена (обучение)')
plt.plot(df.index[time_step:len(train_predict)+time_step], train_predict[:, 0], label='Предсказанная цена (обучение)')
plt.plot(df.index[len(train_predict)+(time_step*2)+1:len(scaled_data)-1], y_test_actual[0], label='Реальная цена (тест)')
plt.plot(df.index[len(train_predict)+(time_step*2)+1:len(scaled_data)-1], test_predict[:, 0], label='Предсказанная цена (тест)')
plt.title(f'Предсказание цены акций {ticker} с помощью LSTM и GA')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.show()

# Анализ расхождений на тестовой выборке
differences = y_test_actual[0] - test_predict[:, 0]

# Построение гистограммы ошибок
plt.figure(figsize=(8, 6))
plt.hist(differences, bins=50)
plt.title('Гистограмма ошибок предсказания (тестовая выборка)')
plt.xlabel('Ошибка (Реальная цена - Предсказанная цена)')
plt.ylabel('Частота')
plt.show()

# Вывод статистики ошибок
mean_error = np.mean(differences)
std_error = np.std(differences)
print(f'Средняя ошибка предсказания: {mean_error}')
print(f'Стандартное отклонение ошибки предсказания: {std_error}')
```
